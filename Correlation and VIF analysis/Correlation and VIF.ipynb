```python
path_1 = "/Users/rajwaghela/Library/CloudStorage/GoogleDrive-rajwaghela4244@gmail.com/My Drive/Thesis/Practical/Data/Vektorabfragen_Updated/VektorabfragenTHLAll1.csv"
path_2 = "/Users/rajwaghela/Library/CloudStorage/GoogleDrive-rajwaghela4244@gmail.com/My Drive/Thesis/Practical/Data/Vektorabfragen_Updated/VektorabfragenTHLAll2.csv"
path_3 = "/Users/rajwaghela/Library/CloudStorage/GoogleDrive-rajwaghela4244@gmail.com/My Drive/Thesis/Practical/Data/Vektorabfragen_Updated/VektorabfragenTHLAll3.csv"

THL_1 = pd.read_csv(path_1)
THL_2 = pd.read_csv(path_2)
THL_3 = pd.read_csv(path_3)
```

```python
# Column abstraction
def col_abstraction(df):
    col_list = []
    for col in df.columns:
        if col.endswith(('_N', 'Abschluss')):
            col_list.append(col)
    return col_list

# grades standardising function
def standard(df):
    for col in df.columns:
        if col.endswith('_N'):
            df[col] = df[col] / 100
    return df
```

```python
BHT_1 = BHT_1[col_abstraction(BHT_1)]
BHT_2 = BHT_2[col_abstraction(BHT_2)]
BHT_3 = BHT_3[col_abstraction(BHT_3)]

BHT_1 = BHT_1.rename(columns={'Abschluss': 'status'})
BHT_2 = BHT_2.rename(columns={'Abschluss': 'status'})
BHT_3 = BHT_3.rename(columns={'Abschluss': 'status'})

BHT_1 = standard(BHT_1)
BHT_2 = standard(BHT_2)
BHT_3 = standard(BHT_3)


THL_1 = THL_1[col_abstraction(THL_1)]
THL_2 = THL_2[col_abstraction(THL_2)]
THL_3 = THL_3[col_abstraction(THL_3)]

THL_1 = THL_1.rename(columns={'Abschluss': 'status'})
THL_2 = THL_2.rename(columns={'Abschluss': 'status'})
THL_3 = THL_3.rename(columns={'Abschluss': 'status'})

THL_1 = standard(THL_1)
THL_2 = standard(THL_2)
THL_3 = standard(THL_3)
```

```python
### swaping values
# after swaping, '1' is -> 'dropout' & '0' is ->'graduate'
BHT_1['status'] = BHT_1['status'] .replace([0, 1], [1, 0])
BHT_2['status'] = BHT_2['status'] .replace([0, 1], [1, 0])
BHT_3['status'] = BHT_3['status'] .replace([0, 1], [1, 0])

### swaping values
# after swaping, '1' is -> 'dropout' & '0' is ->'graduate'
THL_1['status'] = THL_1['status'] .replace([0, 1], [1, 0])
THL_2['status'] = THL_2['status'] .replace([0, 1], [1, 0])
THL_3['status'] = THL_3['status'] .replace([0, 1], [1, 0])
```
